# Transform

```python
Winespectator.com scraping - TRANSFORM
```

```python
# split type column into 3

wineSpect_df[['Province','Variety','Region']] = wineSpect_df.type.str.split("/|:",expand=True) 
wineSpect_df
```

```python
wineSpect_df.columns
```

```python
# drop unnecessary columns
wineSpect_trans = wineSpect_df.drop(['Unnamed: 0', 'type', 'Score','Carneros', 'Napa', 'Santa Barbara', 'Sonoma', 'Anderson Valley',
       'Santa Lucia Highlands', 'Paso Robles', 'Region'], axis=1)
wineSpect_trans
```

```python
wineSpect_trans.columns
```

```python
# check where there are null values
null_columns=wineSpect_trans.columns[wineSpect_trans.isnull().any()]
wineSpect_trans[null_columns].isnull().sum()
```

```python
# drop null values
wineSpect_trans.dropna(inplace=True)
wineSpect_trans
```

```python
# rename columns to match other csv
wineSpect_trans = wineSpect_trans.rename(columns={"Vintage": "Year", "Variety": "Type"})
wineSpect_trans
```

```python
wineSpect_trans.dtypes
```

```python
# reformat year to match other csv
pd.options.display.float_format = '{:.0f}'.format
wineSpect_trans
```

Wine CSV - TRANSFORM

```python
# Create a filtered dataframe from specific columns
wine_data = ["Vintage","Country", "Province","Points","Price","Variety","Title"]
wine_transformed= wine_df[wine_data].copy()

# Rename the column headers
wine_transformed = wine_transformed.rename(columns={"Title": "Name of Wine",
                                                   "Variety":"Type"})
wine_transformed

# Dropping duplicates
# Clean the data by dropping duplicates and setting the index
wine_transformed.drop_duplicates("Name of Wine", inplace=True)
wine_transformed.set_index("Name of Wine", inplace=True)

wine_transformed
```

```python
# Create dataframe for US wines
us_wine_df= wine_transformed.loc[(wine_transformed.Country=="US")]
us_wine_df
```

```python
# Create another column to remove year from Vintage field
us_wine_df['Year']=pd.DatetimeIndex(us_wine_df['Vintage']).year
us_wine_df
```

```python
#Dropping additional columns from df
cleaned_wine_df=us_wine_df.drop(['Vintage','Country'],axis=1)

#sorting wines by price in descending order
cleaned_wine_df.sort_values(['Points'],ascending=False)
cleaned_wine_df
```

MERGE both dataframes

```python
# merge right on 3 columns
merged_df = pd.merge(wineSpect_trans, cleaned_wine_df.reset_index(), on=['Province', 'Type', 'Year'])
merged_df
```

```python
# check to see how many nulls in each column
null_columns=merged_df.columns[merged_df.isnull().any()]
merged_df[null_columns].isnull().sum()
```

```python
# check to find duplicates by name
dupName=merged_df.loc[merged_df.duplicated(subset=["Name of Wine"]), "Name of Wine"].unique().sum()
dupName
```

```python
# drop duplicates
merged_df.drop_duplicates("Name of Wine", inplace=True)
merged_df
```

```python
# drop null values
merged_df.dropna(inplace=True)
merged_df
```

```python
# export to csv
merged_df.to_csv('merged_df.csv') 
```
