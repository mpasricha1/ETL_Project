```python
import pandas as pd
# import dt to convert date to year only
import datetime as dt
import pymongo
```

```python
merged_df = pd.read_csv("merged_df.csv")
merged_df.set_index("Province", inplace = True)
merged_df.head()
```

```python
# # Creating bins to assign Rating definition from scraped data
#     # 95-100 — Classic: Great wines
#     # 90-94 — Outstanding: Wines of superior character and style
#     # 85-89 — Very good: Wines with special qualities
#     # 80-84 — Good: Solid, well-made wines
#     # 75-79 — Mediocre: Drinkable wines that may have minor flaws
#     # 50-74 — Not recommended

bins = [0, 74, 79, 84, 89, 94, 100]
group_names = ["Not Recommended", "Mediocre", "Good", "Very Good", "Outstanding", "Classic"]
merged_df["Rate Definition"] = pd.cut(merged_df["Points"], bins, labels=group_names, include_lowest=True)
merged_df.head()
```

```python
wine_dict = merged_df.to_dict("records")
print(wine_dict)
```

```python
#PYMONGO PUSH
# converted final df to dictionary, reset the index to a string field
# created a DB in Mongo called DataCondasWine_db
# created a collection in Mongo called final_wine_output
# inserted wine_dict to final_wine_output collection in mongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Select database and collection to use (DataCondasWine_db = database name)
db = client.DataCondasWine_db
# Creates a collection in the database and inserts two documents (final_wine_output = collection name)
final_wine_output = db.final_wine_output

#insert dataframe into Mongo, first need to convert to dict
final_wine_output.insert(wine_dict)
```
